# Producción y consumo de energía renovable a nivel mundial 
 * Iniciamos cargando las librerías y el conjunto de datos.

```python
import pandas as pd
import plotly.express as px

# Cargar el conjunto de datos de producción de energía renovable
url1 = 'https://raw.githubusercontent.com/lihkir/Uninorte/main/AppliedStatisticMS/DataVisualizationRPython/Lectures/Python/PythonDataSets/share-of-electricity-production-from-renewable-sources.csv'
df_prod = pd.read_csv(url1)

df_prod
```

* Una vez cargados, procedemos a graficar la producción total de energía a nivel mundial entre 2007 y 2017. El gráfico expone tendencias de aumento en la producción de energía renovable, especialmente suramérica, África con producciones entre 70 y 100%. Algunos países de Europa y Asia también han diversificado su fuente energética, aunque en menor porcentaje, para el período en estudio.

```python
# Ordenar el DataFrame de producción en función de la característica Year
df_prod = df_prod.sort_values('Year')
df_prod = df_prod[(df_prod['Year'] > 2007) & (df_prod['Year'] < 2017)]

# Generar un mapa de coropletas para la producción de energía renovable utilizando el módulo plotly express animado en función del Year
fig_prod = px.choropleth(df_prod, locations="Country", locationmode='country names', color="Renewable electricity (% electricity production)", hover_name="Country", animation_frame="Year", projection="natural earth", title='Producción de energía renovable por país y año', range_color=[0,100])

# Mostrar la figura
fig_prod.show()
```

* Cargamos el segundo conjunto de datos

```python
# Cargar el conjunto de datos de consumo de energía renovable
url2 = 'https://raw.githubusercontent.com/lihkir/Uninorte/main/AppliedStatisticMS/DataVisualizationRPython/Lectures/Python/PythonDataSets/renewable-energy-consumption-by-country.csv'
df_cons = pd.read_csv(url2)
df_cons
```

* En contraste con el mapa anterior, es notorio que los países que más consumen este tipo de energías, se encuentran entre los que menos producen. Esta visualización es fundamental para conocer esta tendencias espaciales y temporales, y apoyar la toma de deciones respecto a la diversificación energétca en diferentes países.

```python
# Ordenar el DataFrame de consumo en base a la característica Year
df_cons = df_cons.sort_values('Year')
df_cons = df_cons[(df_cons['Year'] > 2007) & (df_cons['Year'] < 2017)]

# Generar un mapa coroplético para el consumo de energía renovable utilizando el módulo plotly express animado en función del año
fig_cons = px.choropleth(df_cons, locations="Country", locationmode='country names', color="Total", hover_name="Country", animation_frame="Year", projection="natural earth", title='Consumo de energía renovable por país y año', range_color=[0,400])

# Mostrar la figura
fig_cons.show()
```
